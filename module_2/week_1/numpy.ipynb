```python
# Quest 10
def maxx(x, y):
    if x >= y:
        return x
    else:
        return y

a = np.array([5, 7, 9, 8, 6, 4, 5])
b = np.array([6, 3, 4, 8, 9, 7, 1])

pair_max = np.vectorize(maxx, otypes=[float])
print(pair_max(a, b))
```

```python
# Quest 20
def mapping(loop_val, average_val):
  if loop_val > average_val:
    return "Good"
  elif loop_val == average_val:
    return "Average"
  else:
    return "Bad"

average = np.average(data[:, -1])

result = [mapping(loop_val, average) for loop_val in data[:, -1]]

result[7:10]
```
